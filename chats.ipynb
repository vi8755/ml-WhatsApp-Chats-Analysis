```python
import re
import pandas as pd
```

```python
f=open('WhatsApp Chat with 6th Sem Unofficial.txt','r',encoding='utf-8')
data=f.read()
```

```python
print(data)
```

```python
pattern ='\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s-\s'
```

```python
messages=re.split(pattern,data)[1:]
```

```python
messages
```

```python
dates=re.findall(pattern,data)
dates
```

```python
df=pd.DataFrame({'user_message':messages,'message_date':dates})
#converting message_date type to datetime
df['messages_date']=pd.to_datetime(df['message_date'],format = '%m/%d/%y, %H:%M - ')
df.rename(columns={'message_date':'date'},inplace=True)
df.head()
```

```python
df.shape
```

```python
#seprate user and message
users=[]
messages=[]
for message in df['user_message']:
    entry=re.split('([\w\W]+?):\s',message)
    if entry[1:]: #user name
        users.append(entry[1])
        messages.append(entry[2])
    else:
        users.append('group_notification')
        messages.append(entry[0])
df['users']=users
df['messages']=messages 
df.drop(columns=['user_message'],inplace=True)
df.drop(columns=['date'],inplace=True)
df.sample(10)        
```

```python
df['year']=df['messages_date'].dt.year
```

```python
df.sample()
```

```python
df['month']=df['messages_date'].dt.month_name()
```

```python
df['Date']=df['messages_date'].dt.day #day
```

```python
df['Hour']=df['messages_date'].dt.hour #hour    
```

```python
df['Minute']=df['messages_date'].dt.minute #minute  
```
